```python
df = pd.read_csv("volume_cleaned.csv")
df
```

```python
df['order_direction'].unique()
```

# Incoming orders- Abziehen and leeren

```python
df_in = df[df['order_direction'].isin(['Abziehen','Leeren'])]
```

```python
df_in.shape
```

```python
df_in.to_csv('incoming_orders.csv', index=False, encoding='utf-8-sig')
```
